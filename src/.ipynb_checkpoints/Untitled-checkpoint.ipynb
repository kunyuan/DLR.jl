```julia
ifreq=a[:, 1]; #Matsubara freq
freq = Float64.(freq); #real freq
w = a[:, 2];  #W
```

```julia
kernel = zeros((length(ifreq), length(freq)))
for (i, Ω) in enumerate(ifreq)
    for (j, ω) in enumerate(freq)
        kernel[i, j]= 2ω/(Ω^2+ω^2)
    end
end
```

```julia
coeff = kernel \ w
```

```julia
wp=kernel*coeff
for i in 1:length(ifreq)
    println(ifreq[i], "  ", w[i], "  ", wp[i])
end
```
